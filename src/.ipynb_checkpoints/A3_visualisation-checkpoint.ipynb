## Cell 4 — Data fetch functions

````python
def fetch_violations():
    '''Fetch and flatten A2 violation records from MongoDB.'''
    try:
        client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=3000)
        db = client[MONGO_DB]
        pipeline = [
            {'$unwind': '$violations'},
            {'$project': {
                'car_plate': 1, 'date': 1,
                'violation_type': '$violations.violation_type',
                'camera_id_start': '$violations.camera_id_start',
                'camera_id_end': '$violations.camera_id_end',
                'timestamp_start': '$violations.timestamp_start',
                'timestamp_end': '$violations.timestamp_end',
                'speed_reading': '$violations.speed_reading',
                'risk_score': 1, 'risk_label': 1, 'predicted_at': 1
            }}
        ]
        docs = list(db.violations.aggregate(pipeline))
        client.close()
        if not docs:
            return pd.DataFrame()
        df = pd.DataFrame(docs)
        df['timestamp_start'] = pd.to_datetime(df['timestamp_start'])
        df['timestamp_end']   = pd.to_datetime(df['timestamp_end'])
        df['hour']            = df['timestamp_start'].dt.floor('H')
        df['speed_reading']   = df['speed_reading'].astype(float)
        return df
    except Exception as e:
        print(f'[WARN] MongoDB fetch failed: {e}')
        return pd.DataFrame()




    
def fetch_ml_data():
    """
    Fetch ML-specific collections:
    1. enriched violations with risk_score
    2. risk category samples
    3. all active high-risk watchlist records
    4. proactive alerts
    """

    result = {
        'enriched': pd.DataFrame(),
        'risk_samples': pd.DataFrame(),
        'watchlist': [],
        'alerts': pd.DataFrame()
    }

    try:
        client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=3000)
        db = client[MONGO_DB]

        # ── 1. Fetch ML-scored violation records ─────────────────────────────
        enriched_docs = list(db.violations.aggregate([
            {'$match': {'risk_score': {'$exists': True}}},
            {'$unwind': '$violations'},
            {'$project': {
                'car_plate': 1,
                'date': 1,
                'risk_score': 1,
                'risk_label': 1,
                'predicted_at': 1,

                'camera_id': '$violations.camera_id',
                'camera_id_start': '$violations.camera_id_start',
                'camera_id_end': '$violations.camera_id_end',

                '_id': 0
            }}
        ]))

        if enriched_docs:
            edf = pd.DataFrame(enriched_docs)

            # Convert prediction timestamp into clean hourly buckets.
            edf['predicted_at'] = pd.to_datetime(
                edf['predicted_at'],
                errors='coerce',
                utc=True
            )
            edf['hour'] = edf['predicted_at'].dt.tz_localize(None).dt.floor('H')

            # Ensure risk score is numeric.
            edf['risk_score'] = pd.to_numeric(edf['risk_score'], errors='coerce')

            # Identify which camera detected / produced the scored record.
            def get_predicted_camera(row):
                if 'camera_id' in row and pd.notna(row['camera_id']):
                    return f"Camera {int(row['camera_id'])}"
                elif 'camera_id_end' in row and pd.notna(row['camera_id_end']):
                    return f"Camera {int(row['camera_id_end'])}"
                elif 'camera_id_start' in row and pd.notna(row['camera_id_start']):
                    return f"Camera {int(row['camera_id_start'])}"
                else:
                    return "N/A"

            edf['predicted_camera'] = edf.apply(get_predicted_camera, axis=1)

            # Create simple dashboard-friendly risk categories.
            # This lets us show HIGH / AVERAGE / LOW examples even if the model only stores HIGH_RISK and LOW_RISK.
            edf['risk_category'] = pd.cut(
                edf['risk_score'],
                bins=[-0.01, 0.50, 0.80, 1.01],
                labels=['LOW_RISK', 'AVERAGE_RISK', 'HIGH_RISK']
            )

            result['enriched'] = edf

            # ── meaningful representative records for each risk category ───────────
            # HIGH_RISK: show the 10 highest risk vehicles.
            # AVERAGE_RISK: show 10 vehicles closest to the middle-risk point.
            # LOW_RISK: show the 10 lowest risk vehicles.

            representative_rows = []

            # 1. Top 10 highest-risk vehicles
            high_df = (
                edf[edf['risk_category'] == 'HIGH_RISK']
                .sort_values('risk_score', ascending=False)
                .head(10)
                .copy()
            )
            high_df['display_reason'] = 'Highest risk scores'
            representative_rows.append(high_df)

            # 2. 10 average-risk vehicles closest to the middle of the average range
            average_df = edf[edf['risk_category'] == 'AVERAGE_RISK'].copy()
            if not average_df.empty:
                average_df['distance_from_middle'] = (average_df['risk_score'] - 0.65).abs()
                average_df = (
                    average_df
                    .sort_values('distance_from_middle', ascending=True)
                    .head(10)
                    .copy()
                )
                average_df['display_reason'] = 'Closest to average-risk range'
                representative_rows.append(average_df)

            # 3. Bottom 10 lowest-risk vehicles
            low_df = (
                edf[edf['risk_category'] == 'LOW_RISK']
                .sort_values('risk_score', ascending=True)
                .head(10)
                .copy()
            )
            low_df['display_reason'] = 'Lowest risk scores'
            representative_rows.append(low_df)

            # Combine all representative records into one dashboard table.
            if representative_rows:
                result['risk_samples'] = pd.concat(representative_rows, ignore_index=True)
            else:
                result['risk_samples'] = pd.DataFrame()


        now_iso = datetime.now(timezone.utc).isoformat()

        result['watchlist'] = list(db.watchlist.find(
            {'expires_at': {'$gt': now_iso}},
            {
                'car_plate': 1,
                'risk_score': 1,
                'flagged_at': 1,
                'expires_at': 1,
                'last_seen_date': 1,
                '_id': 0
            }
        ).sort('risk_score', -1).limit(10))

        # ── 3. Fetch proactive alert records ─────────────────────────────────
        alert_docs = list(db.proactive_alerts.find(
            {},
            {
                'car_plate': 1,
                'camera_id': 1,
                'flagged_at': 1,
                'speed_over_limit': 1,
                '_id': 0
            }
        ))

        if alert_docs:
            adf = pd.DataFrame(alert_docs)

            # Convert alert timestamp into clean hourly buckets.
            adf['flagged_at'] = pd.to_datetime(
                adf['flagged_at'],
                errors='coerce',
                utc=True
            )
            adf['hour'] = adf['flagged_at'].dt.tz_localize(None).dt.floor('H')

            result['alerts'] = adf

        client.close()

    except Exception as e:
        print(f'[WARN] ML data fetch failed: {e}')

    return result
````

````python
def get_top20_high_risk_by_camera(ml):
    import pandas as pd

    edf = ml.get("enriched", pd.DataFrame()).copy()

    if edf.empty:
        return edf, pd.DataFrame()

    def get_predicted_camera_id(row):
        if "predicted_camera_id" in row and pd.notna(row["predicted_camera_id"]):
            return int(row["predicted_camera_id"])
        elif "camera_id" in row and pd.notna(row["camera_id"]):
            return int(row["camera_id"])
        elif "camera_id_end" in row and pd.notna(row["camera_id_end"]):
            return int(row["camera_id_end"])
        elif "camera_id_start" in row and pd.notna(row["camera_id_start"]):
            return int(row["camera_id_start"])
        elif "predicted_camera" in row and pd.notna(row["predicted_camera"]):
            text = str(row["predicted_camera"])
            digits = "".join(ch for ch in text if ch.isdigit())
            return int(digits) if digits else None
        else:
            return None

    edf["predicted_camera_id"] = edf.apply(get_predicted_camera_id, axis=1)
    edf = edf[edf["predicted_camera_id"].notna()].copy()

    edf["risk_score"] = pd.to_numeric(edf["risk_score"], errors="coerce")

    # Use risk_category if available, because your bottom HIGH/AVERAGE/LOW table uses this.
    if "risk_category" in edf.columns:
        high_df = edf[
            edf["risk_category"].astype(str).str.upper().str.strip() == "HIGH_RISK"
        ].copy()
    else:
        high_df = edf[
            edf["risk_label"].astype(str).str.upper().str.strip() == "HIGH_RISK"
        ].copy()

    high_df["predicted_at_sort"] = pd.to_datetime(
        high_df["predicted_at"],
        errors="coerce",
        utc=True
    )

    top20 = (
        high_df
        .sort_values(["risk_score", "predicted_at_sort"], ascending=[False, False])
        .head(20)
        .copy()
        .reset_index(drop=True)
    )

    top20["rank"] = top20.index + 1

    return edf, top20


def build_live_prediction_map(ml):
    """
    Build map with one blue circle per camera.
    Circle size = total ML records at that camera.
    Hover shows Top 20 HIGH_RISK count at that camera.
    """
    import os
    import math
    import pandas as pd
    import plotly.graph_objects as go

    possible_camera_paths = [
        "data/camera.csv",
        "../data/camera.csv",
        "./camera.csv"
    ]

    camera_path = None

    for path in possible_camera_paths:
        if os.path.exists(path):
            camera_path = path
            break

    if camera_path is None:
        raise FileNotFoundError(
            "camera.csv not found. Please check data/camera.csv, ../data/camera.csv, or ./camera.csv"
        )

    camera_df = pd.read_csv(camera_path)
    camera_df.columns = [c.lower().strip() for c in camera_df.columns]

    camera_df["camera_id"] = camera_df["camera_id"].astype(int)

    camera_df["camera_name"] = camera_df["camera_id"].apply(lambda x: f"Camera {x}")

    middle_camera = camera_df.iloc[len(camera_df) // 2]

    fig = go.Figure()

    edf, top20 = get_top20_high_risk_by_camera(ml)

    if edf.empty:
        fig.update_layout(
            title="Live Prediction Map — Waiting for ML Risk Scores",
            mapbox=dict(
                style="open-street-map",
                center=dict(
                    lat=float(middle_camera["latitude"]),
                    lon=float(middle_camera["longitude"])
                ),
                zoom=13
            ),
            height=520,
            margin=dict(l=0, r=0, t=50, b=0),
            template="plotly_white"
        )
        return fig

    # Total ML records per camera - used for blue circle radius
    total_counts = (
        edf
        .groupby("predicted_camera_id")
        .size()
        .reset_index(name="total_records")
    )

    # Number of Top 20 HIGH_RISK records per camera
    if not top20.empty:
        top20_counts = (
            top20
            .groupby("predicted_camera_id")
            .size()
            .reset_index(name="top20_high_risk_count")
        )
    else:
        top20_counts = pd.DataFrame(
            columns=["predicted_camera_id", "top20_high_risk_count"]
        )

    # Merge counts into camera locations
    map_df = camera_df.merge(
        total_counts,
        left_on="camera_id",
        right_on="predicted_camera_id",
        how="left"
    )

    map_df = map_df.merge(
        top20_counts,
        left_on="camera_id",
        right_on="predicted_camera_id",
        how="left",
        suffixes=("", "_top20")
    )

    map_df["total_records"] = map_df["total_records"].fillna(0).astype(int)
    map_df["top20_high_risk_count"] = map_df["top20_high_risk_count"].fillna(0).astype(int)

    # Circle size: scaled using sqrt so big values do not become too huge
    map_df["circle_size"] = map_df["total_records"].apply(
        lambda x: 16 if x == 0 else min(70, 18 + math.sqrt(x) * 6)
    )

    fig.add_trace(go.Scattermapbox(
        lat=map_df["latitude"],
        lon=map_df["longitude"],
        mode="markers+text",

        # Show camera number inside the blue circle
        text=map_df["camera_id"].astype(str),
        textposition="middle center",

        marker=dict(
            size=map_df["circle_size"],
            color="#2980b9",
            opacity=0.55
        ),

        name="AWAS Cameras",

        customdata=map_df[[
            "camera_id",
            "camera_name",
            "total_records",
            "top20_high_risk_count",
            "speed_limit",
            "position"
        ]],

        hovertemplate=(
            "<b>%{customdata[1]}</b><br>"
            "Total ML records here: %{customdata[2]}<br>"
            "Top 20 HIGH_RISK records here: %{customdata[3]}<br>"
            "Speed Limit: %{customdata[4]} km/h<br>"
            "Position: %{customdata[5]} km<br>"
            "<extra></extra>"
        )
    ))

    fig.update_layout(
        title="Live Prediction Map — Camera-Level HIGH_RISK Summary",
        mapbox=dict(
            style="open-street-map",
            center=dict(
                lat=float(middle_camera["latitude"]),
                lon=float(middle_camera["longitude"])
            ),
            zoom=13
        ),
        height=520,
        margin=dict(l=0, r=0, t=50, b=0),
        legend=dict(x=0.01, y=0.99, bgcolor="rgba(255,255,255,0.85)"),
        template="plotly_white"
    )

    return fig

````

## Cell 6 — Dashboard Callback

````python
@app.callback(
    [Output('summary-cards',  'children'),
     Output('chart-count',    'figure'),
     Output('insight-count',  'children'),
     Output('chart-speed',    'figure'),
     Output('insight-speed',  'children'),
     Output('chart-camera',   'figure'),
     Output('insight-camera', 'children'),
     Output('chart-scatter',  'figure'),
     Output('insight-scatter','children'),
     Output('chart-risk',     'figure'),
     Output('insight-risk',   'children'),
     Output('chart-alerts',   'figure'),
     Output('insight-alerts', 'children'),
     Output('chart-map',      'figure'),
     Output('alert-panel',    'children'),
     Output('last-updated',   'children')],
    [Input('interval', 'n_intervals')]
)
def update_dashboard(n):
    df     = fetch_violations()
    ml     = fetch_ml_data()
    ts     = f'Last refreshed: {datetime.now().strftime("%H:%M:%S")}'

    def empty_fig(msg='Waiting for stream data...'):
        fig = go.Figure()
        fig.add_annotation(text=msg, xref='paper', yref='paper',
                           x=0.5, y=0.5, showarrow=False,
                           font=dict(size=16, color='#aaa'))
        fig.update_layout(template='plotly_white', height=380)
        return fig

    def card(value, label, color):
        return html.Div([
            html.H2(str(value), style={'margin': '0', 'fontSize': '26px'}),
            html.P(label, style={'margin': '4px 0 0', 'fontSize': '12px'})
        ], style={'textAlign': 'center', 'padding': '14px 22px',
                  'backgroundColor': color, 'color': 'white',
                  'borderRadius': '8px', 'minWidth': '125px'})

    no_data_outputs = (
        [html.Div('No data yet', style={'padding': '20px', 'backgroundColor': '#bdc3c7',
                                        'borderRadius': '8px', 'color': 'white'})],
        empty_fig(), 'No violations yet.',
        empty_fig(), 'No violations yet.',
        empty_fig(), 'No violations yet.',
        empty_fig(), 'No violations yet.',
        empty_fig('Run ml_streaming.ipynb Cell 5b first'), 'ML data not yet available.',
        empty_fig('Run ml_streaming.ipynb Cell 5b first'), 'ML data not yet available.',
        empty_fig('Live prediction map waiting for ML data.'),
        html.P('No watchlist data yet.', style={'color': '#aaa'}),
        ts
    )

    if df.empty:
        return no_data_outputs

    # ── Summary cards ─────────────────────────────────────────────────────────
    total       = len(df)
    inst_count  = len(df[df['violation_type'] == 'INSTANTANEOUS'])
    avg_count   = len(df[df['violation_type'] == 'AVERAGE'])
    mean_speed  = df['speed_reading'].mean()
    unique_cars = df['car_plate'].nunique()
    n_enriched  = len(ml['enriched']) if not ml['enriched'].empty else 0
    n_alerts    = len(ml['alerts']) if not ml['alerts'].empty else 0

    cards = [
        card(f'{total:,}',        'Total Violations',  '#e74c3c'),
        card(f'{inst_count:,}',   'Instantaneous',     '#e67e22'),
        card(f'{avg_count:,}',    'Average Speed',     '#8e44ad'),
        card(f'{mean_speed:.1f}', 'Mean Speed km/h',   '#2980b9'),
        card(f'{unique_cars:,}',  'Unique Vehicles',   '#27ae60'),
        card(f'{n_enriched:,}',   'ML Scored',         '#8e44ad'),
        card(f'{n_alerts:,}',     'Proactive Alerts',  '#e67e22'),
    ]

    styles = {
        'INSTANTANEOUS': dict(color='#e74c3c', dash='solid',  symbol='circle'),
        'AVERAGE':       dict(color='#8e44ad', dash='dash',   symbol='diamond')
    }

    # ── Chart 1: Violation count over time ────────────────────────────────────
    hourly = df.groupby(['hour', 'violation_type']).size().reset_index(name='count')
    fig1   = go.Figure()
    for vtype, s in styles.items():
        data = hourly[hourly['violation_type'] == vtype].sort_values('hour')
        if data.empty: continue
        fig1.add_trace(go.Scatter(
            x=data['hour'], y=data['count'], mode='lines+markers', name=vtype,
            line=dict(color=s['color'], width=2.5, dash=s['dash']),
            marker=dict(size=9, symbol=s['symbol']),
            hovertemplate='%{x|%H:%M}<br>Count: %{y}<extra>' + vtype + '</extra>'
        ))
        mean_c = data['count'].mean()
        mx = data.loc[data['count'].idxmax()]
        fig1.add_annotation(x=mx['hour'], y=mx['count'],
            text=f"MAX: {int(mx['count'])}", showarrow=True, arrowhead=2,
            arrowcolor=s['color'], bgcolor='#fff9c4', bordercolor=s['color'],
            borderwidth=1, font=dict(size=10, color=s['color']))
        if len(data) > 1:
            mn = data.loc[data['count'].idxmin()]
            fig1.add_annotation(x=mn['hour'], y=mn['count'],
                text=f"MIN: {int(mn['count'])}", showarrow=True, arrowhead=2,
                arrowcolor=s['color'], bgcolor='#e3f2fd', bordercolor=s['color'],
                borderwidth=1, font=dict(size=10, color=s['color']))
            for _, sp in data[data['count'] > mean_c * 1.5].iterrows():
                if sp['hour'] != mx['hour']:
                    fig1.add_annotation(x=sp['hour'], y=sp['count'], text='SPIKE',
                        showarrow=True, arrowhead=2, arrowcolor='red',
                        bgcolor='#ffebee', bordercolor='red', borderwidth=1,
                        font=dict(size=9, color='red'))
    fig1.update_layout(title='Chart 1 — Violation Count Over Time: Instantaneous vs Average Speed',
        xaxis_title='Hour of Day', yaxis_title='Number of Violations',
        legend_title='Violation Type', hovermode='x unified',
        template='plotly_white', height=430,
        legend=dict(x=0.01, y=0.99, bgcolor='rgba(255,255,255,0.85)'))
    insight1 = (f'INSTANTANEOUS violations ({inst_count:,}) vs AVERAGE ({avg_count:,}). '
                f'Peaks identify hours of highest enforcement activity. '
                f'SPIKE flags hours >50% above the hourly mean.')

    # ── Chart 2: Average speed over time ──────────────────────────────────────
    hourly_spd = df.groupby(['hour', 'violation_type'])['speed_reading'].mean().reset_index()
    fig2 = go.Figure()
    fig2.add_hline(y=110, line_dash='dot', line_color='#e67e22', line_width=1.5,
                   annotation_text='Cam 1&2 limit: 110 km/h', annotation_position='top right',
                   annotation_font_size=10)
    fig2.add_hline(y=90, line_dash='dot', line_color='#e74c3c', line_width=1.5,
                   annotation_text='Cam 3 limit: 90 km/h', annotation_position='bottom right',
                   annotation_font_size=10)
    for vtype, s in styles.items():
        data = hourly_spd[hourly_spd['violation_type'] == vtype].sort_values('hour')
        if data.empty: continue
        fig2.add_trace(go.Scatter(
            x=data['hour'], y=data['speed_reading'], mode='lines+markers',
            name=f'{vtype} mean speed',
            line=dict(color=s['color'], width=2.5, dash=s['dash']),
            marker=dict(size=9, symbol=s['symbol']),
            hovertemplate='%{x|%H:%M}<br>Avg: %{y:.1f} km/h<extra>' + vtype + '</extra>'
        ))
        mx = data.loc[data['speed_reading'].idxmax()]
        fig2.add_annotation(x=mx['hour'], y=mx['speed_reading'],
            text=f"PEAK: {mx['speed_reading']:.1f} km/h", showarrow=True, arrowhead=2,
            arrowcolor=s['color'], bgcolor='#fff9c4', bordercolor=s['color'],
            borderwidth=1, font=dict(size=10, color=s['color']))
    fig2.update_layout(title='Chart 2 — Average Violation Speed Over Time by Type',
        xaxis_title='Hour of Day', yaxis_title='Average Speed (km/h)',
        hovermode='x unified', template='plotly_white', height=430,
        legend=dict(x=0.01, y=0.99, bgcolor='rgba(255,255,255,0.85)'))
    insight2 = (f'Mean violation speed: {mean_speed:.1f} km/h. '
                f'Hours where average speed rises sharply above reference lines '
                f'indicate periods of aggressive driving.')

    # ── Chart 3: Camera and segment breakdown ─────────────────────────────────
    cam_df = (df[df['violation_type'] == 'INSTANTANEOUS']
              .groupby('camera_id_start').size().reset_index(name='count'))
    cam_df['label'] = cam_df['camera_id_start'].map(
        {1: 'Camera 1 (110)', 2: 'Camera 2 (110)', 3: 'Camera 3 (90)'})
    seg_df = df[df['violation_type'] == 'AVERAGE'].copy()
    seg_df['segment'] = seg_df.apply(
        lambda r: f"Cam {int(r['camera_id_start'])}\u2192{int(r['camera_id_end'])}", axis=1)
    seg_counts = seg_df.groupby('segment').size().reset_index(name='count')
    fig3 = make_subplots(rows=1, cols=2,
        subplot_titles=['Instantaneous by Camera', 'Average Speed by Segment'])
    cam_colors = ['#e74c3c', '#e67e22', '#c0392b']
    for i, row in cam_df.iterrows():
        fig3.add_trace(go.Bar(
            x=[row.get('label', f"Cam {row['camera_id_start']}")],
            y=[row['count']], marker_color=cam_colors[i % 3], showlegend=False,
            text=[f"{row['count']:,}"], textposition='outside'), row=1, col=1)
    for i, row in seg_counts.iterrows():
        fig3.add_trace(go.Bar(
            x=[row['segment']], y=[row['count']],
            marker_color=['#8e44ad', '#6c3483'][i % 2], showlegend=False,
            text=[f"{row['count']:,}"], textposition='outside'), row=1, col=2)
    fig3.update_yaxes(title_text='Violation Count', row=1, col=1)
    fig3.update_yaxes(title_text='Violation Count', row=1, col=2)
    fig3.update_layout(title='Chart 3 — Violation Count by Camera and Segment',
        template='plotly_white', height=400)
    worst_cam = (cam_df.loc[cam_df['count'].idxmax(), 'label']
                 if not cam_df.empty else 'N/A')
    insight3 = (f'{worst_cam} records the highest instantaneous violation count. '
                f'Segment violations reveal drivers who slow at camera points '
                f'but exceed limits between them.')

    # ── Chart 4: Speed scatter with percentile bands ──────────────────────────
    p50 = df['speed_reading'].quantile(0.50)
    p75 = df['speed_reading'].quantile(0.75)
    p90 = df['speed_reading'].quantile(0.90)
    p95 = df['speed_reading'].quantile(0.95)
    fig4 = go.Figure()
    for pval, plabel, pcolor in [
        (p95, f'95th: {p95:.0f} km/h', '#c0392b'),
        (p90, f'90th: {p90:.0f} km/h', '#e67e22'),
        (p75, f'75th: {p75:.0f} km/h', '#f39c12'),
        (p50, f'Median: {p50:.0f} km/h', '#2980b9')
    ]:
        fig4.add_hline(y=pval, line_dash='dash', line_color=pcolor, line_width=1,
                       annotation_text=plabel, annotation_position='right',
                       annotation_font_size=9)
    for vtype, s in styles.items():
        data = df[df['violation_type'] == vtype].sort_values('timestamp_start')
        if data.empty: continue
        fig4.add_trace(go.Scatter(
            x=data['timestamp_start'], y=data['speed_reading'], mode='markers',
            name=vtype, marker=dict(color=s['color'], symbol=s['symbol'], size=6, opacity=0.55),
            hovertemplate='%{x|%H:%M:%S}<br>Speed: %{y:.1f} km/h<extra>' + vtype + '</extra>'
        ))
    max_row = df.loc[df['speed_reading'].idxmax()]
    min_row = df.loc[df['speed_reading'].idxmin()]
    fig4.add_annotation(x=max_row['timestamp_start'], y=max_row['speed_reading'],
        text=f"MAX: {max_row['speed_reading']:.1f} km/h", showarrow=True, arrowhead=2,
        arrowcolor='#c0392b', bgcolor='#c0392b', font=dict(color='white', size=10))
    fig4.add_annotation(x=min_row['timestamp_start'], y=min_row['speed_reading'],
        text=f"MIN: {min_row['speed_reading']:.1f} km/h", showarrow=True, arrowhead=2,
        arrowcolor='#2980b9', bgcolor='#2980b9', font=dict(color='white', size=10))
    spikes = df[df['speed_reading'] > p95].sort_values('speed_reading', ascending=False).head(3)
    for _, sp in spikes.iterrows():
        if sp['timestamp_start'] != max_row['timestamp_start']:
            fig4.add_annotation(x=sp['timestamp_start'], y=sp['speed_reading'],
                text=f"SPIKE: {sp['speed_reading']:.0f}", showarrow=True, arrowhead=2,
                arrowcolor='red', bgcolor='#ffebee', bordercolor='red',
                borderwidth=1, font=dict(size=9, color='red'))
    fig4.update_layout(
        title='Chart 4 — Individual Violation Speeds with Percentile Bands and Spike Annotations',
        xaxis_title='Event Timestamp', yaxis_title='Speed (km/h)',
        template='plotly_white', height=450,
        legend=dict(x=0.01, y=0.99, bgcolor='rgba(255,255,255,0.85)'))
    pct_above_95 = len(df[df['speed_reading'] > p95]) / total * 100
    insight4 = (f'95th percentile: {p95:.1f} km/h. '
                f'{pct_above_95:.1f}% of violations exceed this. '
                f'Median violation speed: {p50:.1f} km/h '
                f'({p50-110:.1f} km/h above the 110 km/h limit).')

    # ── Chart 5 (A3): ML Risk Score Distribution ──────────────────────────────
    # Shows HIGH_RISK vs LOW_RISK violation counts per hour
    # Demonstrates ML layer is enriching violation documents in real time
    edf = ml['enriched']
    if edf.empty:
        fig5 = empty_fig('No ML risk scores yet — run ml_streaming.ipynb Cell 5b')
        insight5 = 'ML scoring not yet active. Run ml_streaming.ipynb Cell 5b first.'
    else:
        risk_hourly = edf.groupby(['hour', 'risk_label']).size().reset_index(name='count')
        fig5 = go.Figure()
        risk_colors = {'HIGH_RISK': '#c0392b', 'LOW_RISK': '#27ae60'}
        for label, color in risk_colors.items():
            data = risk_hourly[risk_hourly['risk_label'] == label].sort_values('hour')
            if data.empty: continue
            fig5.add_trace(go.Bar(
                x=data['hour'], y=data['count'], name=label,
                marker_color=color,
                hovertemplate='%{x|%H:%M}<br>Count: %{y}<extra>' + label + '</extra>'
            ))
        high_count = len(edf[edf['risk_label'] == 'HIGH_RISK'])
        low_count  = len(edf[edf['risk_label'] == 'LOW_RISK'])
        total_scored = len(edf)
        fig5.update_layout(
            title='Chart 5 (A3) — ML Risk Label Distribution Over Time',
            xaxis_title='Hour (predicted_at)', yaxis_title='Violation Count',
            barmode='stack', template='plotly_white', height=400,
            legend=dict(x=0.01, y=0.99, bgcolor='rgba(255,255,255,0.85)'))
        pct_high = high_count / max(total_scored, 1) * 100
        insight5 = (f'ML layer has scored {total_scored:,} violation documents. '
                    f'{high_count:,} ({pct_high:.1f}%) classified as HIGH_RISK. '
                    f'HIGH_RISK plates are written to the watchlist and cross-checked '
                    f'against all subsequent camera events (Nair et al. 2017 '
                    f'predictive monitoring architecture).')

    # ── Chart 6 (A3): Proactive alert volume per hour ─────────────────────────
    # Shows how active the watchlist feedback loop is over time
    adf = ml['alerts']
    if adf.empty:
        fig6 = empty_fig('No proactive alerts yet — watchlist populates after HIGH_RISK plates are scored')
        insight6 = 'No watchlist alerts yet. Let ml_streaming.ipynb run for 5+ minutes.'
    else:
        alert_hourly = adf.groupby('hour').size().reset_index(name='count')
        fig6 = go.Figure()
        fig6.add_trace(go.Bar(
            x=alert_hourly['hour'], y=alert_hourly['count'],
            marker_color='#e67e22', name='Proactive Alerts',
            hovertemplate='%{x|%H:%M}<br>Alerts: %{y}<extra></extra>'
        ))
        # Annotate peak alert hour
        if len(alert_hourly) > 0:
            peak = alert_hourly.loc[alert_hourly['count'].idxmax()]
            fig6.add_annotation(
                x=peak['hour'], y=peak['count'],
                text=f"PEAK: {int(peak['count'])} alerts",
                showarrow=True, arrowhead=2, arrowcolor='#e67e22',
                bgcolor='#fff9c4', bordercolor='#e67e22', borderwidth=1,
                font=dict(size=10, color='#e67e22'))
        fig6.update_layout(
            title='Chart 6 (A3) — Proactive Watchlist Alerts Per Hour',
            xaxis_title='Hour', yaxis_title='Alert Count',
            template='plotly_white', height=380)
        total_alerts = len(adf)
        unique_alert_plates = adf['car_plate'].nunique()
        insight6 = (f'{total_alerts:,} proactive alerts emitted from {unique_alert_plates:,} '
                    f'unique plates. These alerts fired when watchlisted HIGH_RISK plates '
                    f'appeared at any camera BEFORE a formal violation was confirmed by A2 '
                    f'stream-stream joins — demonstrating the ML feedback loop operational value.')

    # ── Alert Panel: Representative risk category records ─────────────────────
    risk_samples = ml.get('risk_samples', pd.DataFrame())


    # ── Representative records from each risk category ───────────────────────
    sample_rows = []

    sample_rows.append(html.Tr([
        html.Th('Risk Category', style={'padding': '8px 12px', 'backgroundColor': '#34495e',
                                        'color': 'white', 'textAlign': 'left'}),
        html.Th('Plate', style={'padding': '8px 12px', 'backgroundColor': '#34495e',
                                'color': 'white', 'textAlign': 'left'}),
        html.Th('Risk Score', style={'padding': '8px 12px', 'backgroundColor': '#34495e',
                                     'color': 'white', 'textAlign': 'center'}),
        html.Th('Predicted Camera', style={'padding': '8px 12px', 'backgroundColor': '#34495e',
                                        'color': 'white', 'textAlign': 'center'}),
        html.Th('Predicted At', style={'padding': '8px 12px', 'backgroundColor': '#34495e',
                                       'color': 'white', 'textAlign': 'left'}),
    ]))

    if risk_samples.empty:
        sample_table = html.P(
            'No ML risk samples available yet. Run ml_streaming.ipynb Cell 5b.',
            style={'color': '#aaa'}
        )
    else:
        for _, row in risk_samples.iterrows():

            category = str(row.get('risk_category', 'N/A'))
            risk = row.get('risk_score', 0)

            if category == 'HIGH_RISK':
                badge_color = '#c0392b'
                bg = '#fff5f5'
            elif category == 'AVERAGE_RISK':
                badge_color = '#e67e22'
                bg = '#fffbf0'
            else:
                badge_color = '#27ae60'
                bg = '#f5fff8'

            predicted_at = row.get('predicted_at', '')
            if pd.notna(predicted_at):
                predicted_at = str(predicted_at)[:19]
            else:
                predicted_at = 'N/A'

            sample_rows.append(html.Tr([
                html.Td(
                    html.Span(category,
                              style={'backgroundColor': badge_color,
                                     'color': 'white',
                                     'padding': '3px 8px',
                                     'borderRadius': '4px',
                                     'fontSize': '12px',
                                     'fontWeight': 'bold'}),
                    style={'padding': '8px 12px'}
                ),
                html.Td(
                    html.Strong(row.get('car_plate', 'N/A')),
                    style={'padding': '8px 12px', 'fontFamily': 'monospace'}
                ),
                html.Td(
                    f"{risk:.3f}",
                    style={'padding': '8px 12px', 'textAlign': 'center',
                           'fontWeight': 'bold'}
                ),

                html.Td(
                    row.get('predicted_camera', 'N/A'),
                    style={'padding': '8px 12px', 'textAlign': 'center',
                        'fontSize': '13px', 'fontWeight': 'bold'}
                ),

                html.Td(
                    predicted_at,
                    style={'padding': '8px 12px', 'fontSize': '12px', 'color': '#555'}
                ),
            ], style={'backgroundColor': bg, 'borderBottom': '1px solid #f0f0f0'}))

        sample_table = html.Table(
            sample_rows,
            style={'width': '100%', 'borderCollapse': 'collapse',
                   'fontSize': '14px', 'marginBottom': '25px'}
        )



    # Combine both parts into the existing alert-panel output.
    alert_panel = html.Div([
        html.H4('Representative Risk Category Records',
                style={'color': '#34495e', 'margin': '0 0 8px 0'}),

        html.P(
            'Displays 10 meaningful representative vehicles from each risk category: highest-risk, average-risk, and lowest-risk records.',
            style={'color': '#777', 'fontSize': '12px'}
        ),

        sample_table


    ])

    # live map: HIGH_RISK predictions points by camera
    fig_map = build_live_prediction_map(ml)
    

    return (cards,
            fig1, insight1,
            fig2, insight2,
            fig3, insight3,
            fig4, insight4,
            fig5, insight5,
            fig6, insight6, fig_map,
            alert_panel, ts)


@app.callback(
    Output("map-click-panel", "children"),
    [Input("chart-map", "clickData"),
     Input("interval", "n_intervals")]
)
def update_map_click_panel(clickData, n):
    ml = fetch_ml_data()
    edf, top20 = get_top20_high_risk_by_camera(ml)


    try:
        camera_id = int(clickData["points"][0]["customdata"][0])
        camera_name = clickData["points"][0]["customdata"][1]
        total_records = int(clickData["points"][0]["customdata"][2])
        top20_count = int(clickData["points"][0]["customdata"][3])
    except Exception:
        return html.P(
            "Please click directly on a camera circle.",
            style={"color": "#777", "margin": "0"}
        )

    camera_top20 = top20[
        top20["predicted_camera_id"].astype(int) == camera_id
    ].copy()

    if camera_top20.empty:
        return html.Div([
            html.H4(
                f"{camera_name}",
                style={"margin": "0 0 8px 0", "color": "#2c3e50"}
            ),
            html.P(
                f"Total ML records at this camera: {total_records}",
                style={"margin": "4px 0", "color": "#555"}
            ),
            html.P(
                "Top 20 HIGH_RISK records here: 0",
                style={"margin": "4px 0", "color": "#c0392b", "fontWeight": "bold"}
            )
        ])

    rows = []

    rows.append(html.Tr([
        html.Th("Rank", style={"padding": "8px", "borderBottom": "1px solid #ddd"}),
        html.Th("Car Plate", style={"padding": "8px", "borderBottom": "1px solid #ddd"}),
        html.Th("Risk Score", style={"padding": "8px", "borderBottom": "1px solid #ddd"}),
        html.Th("Predicted At", style={"padding": "8px", "borderBottom": "1px solid #ddd"})
    ]))

    for _, row in camera_top20.iterrows():
        predicted_at = row.get("predicted_at", "N/A")

        rows.append(html.Tr([
            html.Td(row.get("rank", "N/A"), style={"padding": "8px", "textAlign": "center"}),
            html.Td(row.get("car_plate", "N/A"), style={"padding": "8px", "fontWeight": "bold"}),
            html.Td(f"{row.get('risk_score', 0):.3f}", style={"padding": "8px", "textAlign": "center"}),
            html.Td(str(predicted_at), style={"padding": "8px", "fontSize": "12px", "color": "#555"})
        ], style={"borderBottom": "1px solid #eee"}))

    return html.Div([
        html.H4(
            f"{camera_name}",
            style={"margin": "0 0 8px 0", "color": "#2c3e50"}
        ),

        html.P(
            f"Total ML records at this camera: {total_records}",
            style={"margin": "4px 0", "color": "#555"}
        ),

        html.P(
            f"Top 20 HIGH_RISK records here: {top20_count}",
            style={"margin": "4px 0 12px 0", "color": "#c0392b", "fontWeight": "bold"}
        ),

        html.Table(
            rows,
            style={
                "width": "100%",
                "borderCollapse": "collapse",
                "fontSize": "13px",
                "backgroundColor": "white"
            }
        )
    ])


print("Callback defined — run Cell 7 to start the server")
````
